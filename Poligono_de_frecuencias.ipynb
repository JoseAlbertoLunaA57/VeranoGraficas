# Grafica Poligono de frecuencias
- [√çndice](index.ipynb)

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
frecuencias = [5, 15, 22, 30, 10]  # Frecuencias de cada clase
clases = ['Clase 1', 'Clase 2', 'Clase 3', 'Clase 4', 'Clase 5']  # Nombres de las clases


# Calcular los puntos x para el pol√≠gono de frecuencias
x = [i for i in range(len(frecuencias))]

# Crear el pol√≠gono de frecuencias
plt.plot(x, frecuencias, marker='o', linestyle='-', color='b')

# A√±adir etiquetas y t√≠tulo
plt.xticks(x, clases)  # Etiquetas en el eje x
plt.xlabel('Clases')
plt.ylabel('Frecuencias')
plt.title('Pol√≠gono de Frecuencias')

# Mostrar la gr√°fica
plt.grid()
plt.show()
```
