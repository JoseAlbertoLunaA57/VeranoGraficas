```python
import matplotlib.pyplot as plt

# Datos para el diagrama de barras
categorias = ['Categor√≠a A', 'Categor√≠a B', 'Categor√≠a C', 'Categor√≠a D', 'Categor√≠a E', 'Categor√≠a F']
valores = [10, 15, 7, 20, 25, 30]

# Crear el diagrama de barras horizontales
plt.barh(categorias, valores, color='skyblue')

# A√±adir t√≠tulo y etiquetas
plt.title('Diagrama de Barras Horizontales')
plt.xlabel('Valores')
plt.ylabel('Categor√≠as')

# Mostrar la gr√°fica
plt.show()
```
