```python
import matplotlib.pyplot as plt

# Datos para el diagrama de barras
categorias = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D', 'Categoría E', 'Categoría F']
valores = [10, 15, 7, 20, 25, 30]

# Crear el diagrama de barras horizontales
plt.barh(categorias, valores, color='skyblue')

# Añadir título y etiquetas
plt.title('Diagrama de Barras Horizontales')
plt.xlabel('Valores')
plt.ylabel('Categorías')

# Mostrar la gráfica
plt.show()
```
