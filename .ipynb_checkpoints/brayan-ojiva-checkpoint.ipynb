```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 6]

# Calcular la frecuencia acumulativa
values, base = np.histogram(data, bins=range(1, 8))
cumulative = np.cumsum(values)

# Graficar la ojiva
plt.figure(figsize=(8, 5))
plt.plot(base[1:], cumulative, marker='o', linestyle='-')
plt.title('Gráfico de Ojiva')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Acumulativa')
plt.grid()
plt.xticks(range(1, 8))
plt.show()
```
